```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/lilieven/Documents/Master/Dataliteracy/testlizenz-tuebingen-dataliteracy4students_1701258214900/testlizenz-tuebingen-dataliteracy4students_1701258214900.csv", sep=";", encoding='unicode_escape', decimal=",")
#df['anz_zimmer'] = df['anz_zimmer'].replace(',', '.', regex=True).astype(float)
#df['kstn_miete_gesamt_pqm'] = df['kstn_miete_gesamt_pqm'].replace(',', '.', regex=True).astype(float)
df.info(verbose=True)
```

```python
df_copy = pd.DataFrame()

for col in df.columns:
    allowed= ["int64", "float64"]
    if str(df[col].dtype) in allowed:
        df_copy[col] = df[col]

df_copy.info(verbose=True)
```

```python
df = df[df['anz_zimmer'] != '<NA>']
df = df[df['kstn_miete_gesamt_pqm'] != '<NA>']
types_of_heating = ['aus_heiz_alt_erdw_janein', 'aus_heiz_alt_solar_janein', 'aus_heiz_alt_pellet_janein', 'aus_heiz_alt_pumpe_janein', 'aus_heiz_blockkw_janein', 'aus_heiz_fernwaerme_janein', 'aus_heiz_fussboden_janein', 'aus_heiz_keine_janein', 'aus_kamin_janein']


for heating_type in types_of_heating:
    sns.lmplot(x="anz_zimmer", y="kstn_miete_gesamt_pqm", hue=heating_type, data=df)
    plt.title(f"LM Plot for {heating_type}")
    plt.show()

 # Count the number of flats with the specified heating type
for heating_type in types_of_heating:
    count_of_flats = df[heating_type].sum() 
    print('Number of flats with ' + heating_type + ': ' + str(count_of_flats))
```

```python
df = df[df['baujahr'] != '<NA>']
for heating_type in types_of_heating:
    sns.lmplot(x="baujahr", y="kstn_miete_gesamt_pqm", hue=heating_type, data=df)
    plt.title(f"LM Plot for {heating_type}")
    plt.show()

```
